```python
import cv2
import matplotlib.pyplot as plt

#resimleri al
img1=cv2.imread("img/manzara3.jpg")
#bgr to rgb
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img2=cv2.imread("img/manzara4.jpg")
#bgr to rgb
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(img1)

plt.figure()
plt.imshow(img2)


#resimleri birlestirmeden nce boyutlarna bakalm
print(img1.shape)
print(img2.shape)

#boyutlarn dzenleyelim
img1=cv2.resize(img1, (600,600))
print(img1.shape)

img2=cv2.resize(img2, (600,600))
print(img2.shape)

plt.figure()
plt.imshow(img1)

plt.figure()
plt.imshow(img2)

#kartrlm resim =alpha*img1+ beta*img2 seklinde olacaktr
blended=cv2.addWeighted(src1= img1, alpha=0.5, src2= img2, beta = 0.5, gamma=0)
plt.figure()
plt.imshow(blended)
```
