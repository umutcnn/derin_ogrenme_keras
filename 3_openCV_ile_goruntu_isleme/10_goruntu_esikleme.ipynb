```python
import cv2
import matplotlib.pyplot as plt


#resimi al
img=cv2.imread("img/manzara3.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")

#thresh değeri üstündeki değerleri beyaz, maxval değerlerini ise siyah yapar
#eşiklemeyi yapalım
_,thresh_img= cv2.threshold(img, thresh = 65, maxval=255, type=cv2.THRESH_BINARY)
plt.figure()
plt.imshow(thresh_img,cmap="gray")

#aydınlatmalarına göre threshold nasıl uygulanır
thresh_img2=cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11,8)
plt.figure()
plt.imshow(thresh_img2,cmap="gray")
```
