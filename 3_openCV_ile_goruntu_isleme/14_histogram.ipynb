# Histogram
Görüntü histogramı, dijital görüntüdeki ton dağılımının grafiksel bir temsili olarak işlev gören bir histogram türüdür Her bir ton değeri için piksel sayısını içerir.
Belirli bir görüntü için histograma bakılarak, ton dağılımı alınabilir

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("img/histogram.jpg")
img_vis=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(),plt.imshow(img_vis),plt.axis("off"),plt.title("orijinal"),plt.show()
```

```python
img_hist=cv2.calcHist([img_vis], channels=[0],mask=None, histSize=[256], ranges=[0,256])
plt.figure(),plt.plot(img_hist)
```

```python
#renk ayrımı alarak histogram cizdirme
color = ("b","g","r")
plt.figure()
for i,c in enumerate(color):
    hist = cv2.calcHist([img], channels=[i],mask=None, histSize=[256], ranges=[0,256])
    plt.plot(hist,color = c)
plt.title('HISTOGRAM')
```

```python
golden_gate=cv2.imread("img/golden_gate.jpg")
golden_gate_vis=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(),plt.imshow(img_vis),plt.axis("off"),plt.title("golden_gate"),plt.show()

mask=np.zeros(golden_gate.shape[:2], np.uint8)
plt.figure(),plt.imshow(mask,cmap="gray"),plt.axis("off"),plt.title("maskeli"),plt.show()

mask[100:500,200:500]=255
plt.figure(),plt.imshow(mask,cmap="gray"),plt.axis("off"),plt.title("maskeli"),plt.show()

masked_img_vis = cv2.bitwise_and(golden_gate_vis,golden_gate_vis,mask=mask)
plt.figure(),plt.imshow(masked_img_vis,cmap="gray"),plt.axis("off"),plt.title("maskeli"),plt.show()

masked_img = cv2.bitwise_and(golden_gate,golden_gate,mask=mask)
masked_img_hist=cv2.calcHist([golden_gate], channels=[0],mask=None, histSize=[256], ranges=[0,256])
plt.figure(),plt.plot(masked_img_hist)
```

```python
#histogram eşitleme yöntemi
#karşıtlık(kontrast) arttırma
img_equ = cv2.imread("img/histogram2.jpg",0)
plt.figure(),plt.imshow(img_equ,cmap="gray"),plt.title("Original"),plt.show()




eq_hist = cv2.equalizeHist(img_equ)
plt.figure(),plt.imshow(eq_hist,cmap="gray"),plt.title("Contrast"),plt.show()

img_hist=cv2.calcHist([img_equ], channels=[0],mask=None, histSize=[256], ranges=[0,256])
plt.figure(),plt.plot(img_hist)

img_hist=cv2.calcHist([eq_hist], channels=[0],mask=None, histSize=[256], ranges=[0,256])
plt.figure(),plt.plot(img_hist)
```
