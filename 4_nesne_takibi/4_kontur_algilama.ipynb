```python
#kütüphaneleri import edelim
import cv2
import matplotlib.pyplot as plt
import numpy as np

#resmi al 
img = cv2.imread("img/contour.jpg", 0)
plt.figure(),plt.imshow(img,cmap="gray")
```

```python
#resmime contour uyguladım
contours , hierarchy = cv2.findContours(img,cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

external_contour = np.zeros(img.shape)
internal_contour = np.zeros(img.shape)


for i in range(len(contours)):
    
    #icini beyaz yapiyorum
    if hierarchy[0][i][3]== -1:
        cv2.drawContours(external_contour,contours,i,255,-1)
        
    #resmin distaki konturunu yakalamaya calisiyorum
    else:
        cv2.drawContours(internal_contour,contours,i,255,-1)
        
#gorsellestir
plt.figure(),plt.imshow(external_contour,cmap="gray")
plt.figure(),plt.imshow(internal_contour,cmap="gray")
```
