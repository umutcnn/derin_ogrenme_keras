```python
erkek = cv2.imread('img/erkek.jpg',0)
plt.imshow(erkek,cmap='gray')
```

# İnsanların yüzlerini bulalım

```python
#eğitilmiş haarcascade içeri alalım
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    #gelen fotomun bir kopyasını oluşturuyorum
    face_img = img.copy()
    
    face_recs = face_cascade.detectMultiScale(face_img)
    
    #çizelim
    for (x,y,w,h) in face_recs:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,255),10)
    return face_img
```

```python
result=detect_face(erkek)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result2=detect_face(kadin)
plt.imshow(result2,cmap='gray')
```

```python
result3=detect_face(toplu_foto)
plt.imshow(result3,cmap='gray')
```

# insanların gözlerini bulalım

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eye(img):
    
    #gelen fotomun bir kopyasını oluşturuyorum
    face_img = img.copy()
    
    eyes = eye_cascade.detectMultiScale(face_img)
    
    #çizelim
    for (x,y,w,h) in eyes:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,255),3)
    return face_img
```

```python
result4=detect_eye(erkek)
plt.imshow(result4,cmap='gray')
```

```python
result5=detect_eye(kadin)
plt.imshow(result5,cmap='gray')
```
