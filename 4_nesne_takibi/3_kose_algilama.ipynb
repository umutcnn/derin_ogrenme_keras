# Köşe algılama
• Özellik Nedir?
- Bilgisayarla görmede, genellikle bir resmin/videonun farklı çerçeveleri arasında eşleşen noktalar bulmamız gerekir. Çünkü, iki görüntünün birbiriyle nasıl ilişkili olduğunu bilirsek, her iki görüntüyü de bilgi almak için kullanabiliriz.
- Eşleştirme noktaları dediğimizde, genel anlamda sahnedeki kolayca - tanıyabileceğimiz özelliklere atıfta bulunuyoruz. Bu özelliklere özellikler diyoruz.
Bu özellikler benzersiz bir şekilde tanınabilir olmalıdır.
• Temel özellikler nelerdir?
- Kenarlar
- Köşeler: 
    Köşeler, iki kenarın kesişimi olduğu için bu iki kenarın yönlerinin değiştiği bir noktayı temsil eder.
    Köşeler, resimdeki renk geçişindeki bir varyasyonu temsil ettiğinden, bu "varyasyonu" arayacağız. Görüntü yoğunluğundaki         varyasyonu arayacağız.

```python
#kütüphaneleri import edelim
import cv2
import matplotlib.pyplot as plt
import numpy as np

#resmi al 
img = cv2.imread("img/sudoku.png", 0)
img=np.float32(img)
print(img.shape)
plt.figure(),plt.imshow(img,cmap="gray")
```

```python
#harris corner detection
dst = cv2.cornerHarris(img,blockSize=2, ksize=3,k=0.04)
plt.figure(),plt.imshow(dst,cmap="gray")
```

```python
dst=cv2.dilate(dst,None)
img[dst>0.2*dst.max()]=1
plt.figure(),plt.imshow(dst,cmap="gray"),plt.show()
```

```python
#shi tomasi corner detection
img = cv2.imread("img/sudoku.png", 0)
img=np.float32(img)
corners = cv2.goodFeaturesToTrack(img, 120, 0.01, 10)
corners=np.int64(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img, (x,y),3,(125,125,125),cv2.FILLED)
plt.figure(),plt.imshow(img,cmap="gray"),plt.show()
```
